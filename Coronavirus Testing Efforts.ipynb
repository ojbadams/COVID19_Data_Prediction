To account for the variations in number of tests a day this will look into the percentage of positive cases out of number of tests carried out. 

```python
from pandas import * 
from datetime import *
import matplotlib.pyplot as plt
```

```python
df = read_csv("covid-19-tests-uk.csv")
df = df[["Date", "DailyTestsPerformed", "DailyPeopleTested", "DailyPositive"]]
df["Date"] = to_datetime(df["Date"])

df.dropna(subset = ["DailyTestsPerformed", "DailyPeopleTested", "DailyPositive"], how = "all", inplace = True)
df.set_index("Date", inplace = True)
df
```

Forward fill of NaN values

```python
df.fillna(method = "ffill", inplace = True)
df
```

```python
df["PerPositive"] = df["DailyPositive"]/df["DailyPeopleTested"]
df["PerPositiveRoll"] = df["PerPositive"].rolling("7D", min_periods = 1).mean()
```

```python
df["PerPositiveRoll"].plot(figsize = (10,3))
```

```python
plt.figure(figsize = (10,3))
plt.plot(df.index, df["PerPositiveRoll"], label = "Positive Percentage Rolling Av.")
plt.plot(df.index, df["PerPositive"], label = "Positive Percentage")
plt.title("Percentage of Positive Cases Divided By People Tested")
plt.xlabel("Date")
plt.ylabel("Percentage of Positive Cases")
plt.legend()
```

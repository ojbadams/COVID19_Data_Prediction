This is a brief look into trends of daily deaths and cases, concerning the current COVID-19 pandemic.

Data was sought from the website "Our World in Data" (https://ourworldindata.org/coronavirus). Only data from the UK was analysed. 

```python
from pandas import * 
import matplotlib.pyplot as plt 
import numpy as np 
from datetime import *
```

```python
df = read_csv("owid-covid-data.csv")
dfUK = df[df['location'] == "United Kingdom"]
dfUK = dfUK[["date", "total_cases", "new_cases", "total_deaths", "new_deaths"]]
dfUK["date"] = to_datetime(dfUK["date"])


dfUK = dfUK.set_index("date")
dfUK
```

The cell below informs us that there is no missing data within the dataset. 

```python
dfUK.isnull().sum()
```

```python
dfUK[["new_cases"]].plot(figsize = (10,6), grid = True, title = "Total Covid-19 Cases Jan - May", xlim = [datetime(2020,4,20), datetime(2020,5,28)])
```

Looking at the graph above, there is a negative value among the new daily cases. 

From the cell below, this value is given to be -525. It is possible that this is a value that has been used to correct previous numbers of new cases. It is also possible that it is erroneous. As there appears to be a downward trend, the mean of the two values either side will be used to correct this value. 

The DataFrame was checked to ensure that there were no other values less than zero and this was satisfied. 

```python
dfUK["new_cases"].loc[datetime(2020,5,21)]
```

```python
(dfUK < 0).sum()
```

```python
correctionMean = (dfUK["new_cases"].loc[datetime(2020,5,20)]+dfUK["new_cases"].loc[datetime(2020,5,22)])/2
correctionMean
```

```python
dfUK["new_cases"][datetime(2020,5,21)] = correctionMean

#Much better, see graph below. 
```

Graphs were produced of daily new cases and daily new deaths 

```python
dfUK[["new_cases"]].plot(figsize = (10,6), grid = True, title = "Daily New COVID-19 Cases Jan - May", xlim = [datetime(2020,1,1), datetime(2020,5,22)])
```

```python
dfUK[["new_deaths"]].plot(figsize = (10,6), grid = True, title = "Daily New COVID-19 Deaths Jan - May", xlim = [datetime(2020,1,1), datetime(2020,5,22)])
```

In both graphs while there is a clear downward trend it is difficult to apprehend the average decreases due to spikes and dips in the data. Due to this, it is useful to calculate the rolling average over a 7 day period and this is performed below. 

```python
dfUK["Rolling Average Deaths"] = dfUK["new_deaths"].rolling("7D", min_periods = 1).mean()
dfUK["Rolling Average Cases"] = dfUK["new_cases"].rolling("7D", min_periods = 1).mean()
```

```python
dfUK[["new_cases", "Rolling Average Cases"]].plot(figsize =(10,6), grid = True, title = "Daily New COVID-19 Cases UK")
```

```python
dfUK[["new_deaths", "Rolling Average Deaths"]].plot(figsize = (10,6), grid = True, title = "Daily New COVID-19 Deaths UK")
```

In both graphs it is now indisputable that there exists a downwards trend. The rolling average is also useful as now the peak in both graphs is easier to identify. 

The rolling average is also useful as it can be used in estimation of the continuation of the peak. This is explored further in the next section. 

# Estimation of Predicted Daily Deaths

In this section, Mathematica was used in conjunction with Python. This is becuase of some of the better resources avaliable to Mathematica in the form of the "Fit" function. 

The Fit function will find a function of a specified degree to fit to the relevant data. In this case, a polynomial of degree 2 was used. Different functions can be used for this purpose, but a quadratic appeared to be suitable for a simple approximation. 

As the graph is not symmetrical, a subset of the data was specified for use with the Fit function, specifically that from the peak onwards. 

Lastly, the data input is somewhat different in Mathematica, the following cell is used to "translate" a dataframe to code that can be inputed in Mathematica. For the sake of this, the datetime index has been simplified to a basic domain of 1,2,..., etc. 

The Mathematica notebook can be found attatched. 

```python
val = 1
peakVal = 110
outStr = "{"
for i in dfUK["Rolling Average Deaths"].iloc[peakVal:]: 
    outStr += "{" + str(val) + "," + str(i) + "},"
    val += 1 
    
outStr
```

```python
dfUK.iloc[peakVal]
```

The final function was copied below and then converted so it could be plotted against the real data. 

Lastly, Mathematica calculated that the number of deaths would turn negative at day 65.7308 and thus the domain was restricted to an upper limit of 66 days, or the 24th June 2020. 

```python
zeroDeathDay = 67 

datetime(2020,4,18) + timedelta(67)
```

```python

xPred = []
yPred = []

for tVal in range(1,zeroDeathDay): 
    xPred.append(datetime(2020,4,18) + timedelta(tVal)) 
    yPred.append(962.24 - 22.6839*tVal + 0.12239*tVal**2)
    tVal += 1 

xRe = dfUK.index
yRe = dfUK["Rolling Average Deaths"]

plt.figure(figsize = (10,7))
plt.plot(xPred, yPred, "--", label = "Predicted daily COVID-19 Deaths")
plt.plot(xRe,yRe, label = "Current daily COVID-19 Deaths")
plt.xlabel("Date")
plt.ylabel("No. of Deaths")
plt.title("Predicted and Actual Daily COVID-19 Deaths")
plt.legend()
plt.show()
```

As seen above, assuming the data continues its current trajectory, the number of deaths will continue to decrease and reach zero in roughly a month. 

It would seem that the data fits the graph well. There is a liklihood that the deaths may in fact start to plateau as the number of deaths decrease, but this is not possible to represent with a polynomial of degree two.  

As previously mentioned, this model estimates we'll reach zero daily COVID-19 deaths on roughly June 24th

Lastly, an estimation to the remainding number of total COVID-19 deaths can be predicted by calculating the integration between the end of the data and the point that deaths are estimated to reach zero. Mathematica is very useful for this, and has estimated 4622.85 more deaths.

```python
val = 1
peakVal = datetime(2020,4,15)
outStr = "{"
for i in dfUK["Rolling Average Cases"].loc[peakVal:]: 
    outStr += "{" + str(val) + "," + str(i) + "},"
    val += 1 
    
outStr
```

```python
zeroCasesDay = 57 

datetime(2020,4,15) + timedelta(zeroCasesDay)
```

```python

xPred = []
yPred = []

for tVal in range(1,zeroCasesDay): 
    xPred.append(datetime(2020,4,18) + timedelta(tVal)) 
    yPred.append(5119.55 + 23.1877*tVal - 1.98463*tVal**2)
    tVal += 1 

xRe = dfUK.index
yRe = dfUK["Rolling Average Cases"]

plt.figure(figsize = (10,7))
plt.plot(xPred, yPred, "--", label = "Predicted Covid-19 Cases")
plt.plot(xRe,yRe, label = "Current Covid-19 Cases")
plt.xlabel("Date")
plt.ylabel("No. of Cases")
plt.title("Predicted and Actual Covid-19 Cases")
plt.legend()
plt.show()
```

It would appear that this function does not map the curve as well. However, the function estimates we'll reach zero new cases on the roughly 11th of June. Remarkably, this works well with the estimation of number of deaths new which reaches zero in about two weeks time after this.
